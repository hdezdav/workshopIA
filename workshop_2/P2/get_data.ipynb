```python
import os
import kagglehub

# download latest version
path = kagglehub.dataset_download("jangedoo/utkface-new")

print("path to dataset files:", path)
```

```python

# Usar la ruta obtenida de kagglehub en lugar de una ruta hardcodeada
for folder in ["utkface_aligned_cropped", "crop_part1", "UTKFace"]:
    folder_path = os.path.join(path, folder)
    # filter for actual image files to get an accurate count and sample
    if os.path.exists(folder_path):
        images = [f for f in os.listdir(folder_path) if f.lower().endswith((".jpg", ".jpeg", ".png"))]
        print(f"{folder}: {len(images)} images")
        print("sample:", images[:5])
    else:
        print(f"{folder}: carpeta no encontrada en {folder_path}")
    print("-" * 40)
```
